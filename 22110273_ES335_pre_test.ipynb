Answer 01:

Matrix multiplication involves computing the dot product of each row of the first matrix with each column of the second matrix. Let's consider two matrices: A with dimensions (n, k) and B with dimensions (k, m).

For each element C[i, j] in the resulting matrix C (product of A and B), it is computed as the sum of the products of corresponding elements from the ith row of A and the jth column of B.

The number of multiplications needed to compute each element C[i, j] is equal to the number of elements in a row of A (which is k). The number of additions needed is one less than the number of multiplications (k-1), as you're summing up the products.

For a single element C[i, j], you need k multiplications and (k-1) additions. Since there are n rows and m columns in the resulting matrix C, the total number of multiplications and additions can be calculated as follows:

Number of multiplications = n * m * k
Number of additions = n * m * (k - 1)

So, the total number of multiplications required is n * m * k, and the total number of additions required is n * m * (k - 1) when multiplying a (n, k) matrix by a (k, m) matrix.

```python
def multiply_matrices_list_of_lists(A, B):
    # dimensions of matrices A and B
    n = len(A)
    k = len(A[0])
    m = len(B[0])

    # Initialize
    C = [[0 for _ in range(m)] for _ in range(n)]

    # matrix multiplication
    for i in range(n):
        for j in range(m):
            for p in range(k):
                C[i][j] += A[i][p] * B[p][j]

    return C

import numpy as np

def multiply_matrices_numpy(A, B):
    return np.dot(A, B)

import time

# matrices A and B
A = [[1, 2], [3, 4], [5, 6]]
B = [[7, 8], [9, 10]]

# list of lists
start_time_list_of_lists = time.time()
result_list_of_lists = multiply_matrices_list_of_lists(A, B)
end_time_list_of_lists = time.time()
time_list_of_lists = end_time_list_of_lists - start_time_list_of_lists


A_np = np.array(A)
B_np = np.array(B)
start_time_numpy = time.time()
result_numpy = multiply_matrices_numpy(A_np, B_np)
end_time_numpy = time.time()
time_numpy = end_time_numpy - start_time_numpy

# results and timings
print("Result using list of lists:")
print(result_list_of_lists)
print("Time taken using list of lists:", time_list_of_lists)

print("\nResult using NumPy:")
print(result_numpy)
print("Time taken using NumPy:", time_numpy)
```

```python
import time

def findMedian1(arr):
    n = len(arr)

    for _ in range(n // 2 - 1):
        max_val = arr[0]
        for i in range(len(arr)):
            if arr[i] > max_val:
                max_val = arr[i]
        arr.remove(max_val)

    max1 = arr[0]
    for i in range(len(arr)):
        if arr[i] > max1:
            max1 = arr[i]

    arr.remove(max1)
    max2 = 0
    for i in range(len(arr)):
        if arr[i] > max2:
            max2 = arr[i]

    if n % 2 != 0:
        median = max2
    else:
        median = (max1 + max2) / 2

    return median

np_arr = range(1, 1000)
arr = list(np_arr)

start_time = time.time()
median = findMedian1(arr)
end_time = time.time()

print(f"The Median is {median} by one pass method.")
print(f"Time taken by the above method is {end_time - start_time} seconds")
```

```python
import jax
import jax.numpy as jnp

def f(x, y):
    return x**2 * y + y**3 * jnp.sin(x)

# Use JAX
grad_f = jax.grad(f, argnums=(0, 1))

random_values = [(1.5, 2.0), (0.8, 1.2), (-1.0, 3.0)]

for x_val, y_val in random_values:
    gradient_result = grad_f(x_val, y_val)
    gradient_x, gradient_y = gradient_result
    print(f"Gradients at (x, y) = ({x_val}, {y_val}):")
    print(f"   Gradient with respect to x: {gradient_x}")
    print(f"   Gradient with respect to y: {gradient_y}")
    print()
```

```python
class Student:
    def __init__(self, roll_number, name, marks):
        self.roll_number = roll_number
        self.name = name
        self.marks = marks

class Branch:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

class Year:
    def __init__(self, year):
        self.year = year
        self.branches = []

    def add_branch(self, branch):
        self.branches.append(branch)

# Creating an overall database as a list of Year objects
students_database = []

# Adding data for the year 2022
year_2022 = Year(2022)

branch_1_2022 = Branch('Branch 1')
branch_1_2022.add_student(Student(1, 'N', {'Maths': 100, 'English': 70}))

branch_2_2022 = Branch('Branch 2')

year_2022.add_branch(branch_1_2022)
year_2022.add_branch(branch_2_2022)

students_database.append(year_2022)

# Adding data for other years, branches, and students follows a similar pattern

# Example: Adding data for the year 2023
year_2023 = Year(2023)
branch_1_2023 = Branch('Branch 1')
branch_2_2023 = Branch('Branch 2')
year_2023.add_branch(branch_1_2023)
year_2023.add_branch(branch_2_2023)
students_database.append(year_2023)

# Displaying the overall database
for year in students_database:
    print(f"\nYear {year.year}:")
    for branch in year.branches:
        print(f"  {branch.name}:")
        for student in branch.students:
            print(f"    Roll Number: {student.roll_number}, Name: {student.name}, Marks: {student.marks}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_function(ax, x_values, y_values, function_name):
    ax.plot(x_values, y_values, label=f'y = {function_name}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f'Plot of y = {function_name}')
    ax.legend()

# Define the domain
x_values = np.arange(0.5, 100.5, 0.5)

fig, axes = plt.subplots(3, 3, figsize=(15, 15))

axes = axes.flatten()

# Plotting
plot_function(axes[0], x_values, x_values, 'x')
plot_function(axes[1], x_values, x_values**2, 'x^2')
plot_function(axes[2], x_values, (x_values**3) / 100, 'x^3 / 100')
plot_function(axes[3], x_values, np.sin(x_values), 'sin(x)')
plot_function(axes[4], x_values, np.sin(x_values) / x_values, 'sin(x) / x')
plot_function(axes[5], x_values, np.log(x_values), 'log(x)')
plot_function(axes[6], x_values, np.exp(x_values), 'e^x')

# Hide remaining empty subplots
for i in range(7, 9):
    fig.delaxes(axes[i])

# Adjust layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import numpy as np

def custom_argmin(arr):
    # Check if the array is empty
    if len(arr) == 0:
        raise ValueError("Input array is empty.")

    min_value = arr[0]
    min_index = 0

    for i, value in enumerate(arr):
        if value < min_value:
            min_value = value
            min_index = i

    return min_index

array_example = np.array([4, 2, 8, 1, 5])
result_custom_argmin = custom_argmin(array_example)

result_np_argmin = np.argmin(array_example)

print("Array:", array_example)
print("Custom argmin result:", result_custom_argmin)
print("np.argmin result:", result_np_argmin)
```
